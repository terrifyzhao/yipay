```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
%matplotlib inline
pd.options.mode.chained_assignment = None  # default='warn'
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=15)


train_base = pd.read_csv('data/train_base.csv')

train_op = pd.read_csv('data/train_op.csv')

train_trans = pd.read_csv('data/train_trans.csv')

train_label = pd.read_csv('data/train_label.csv')
```

```python
df = pd.merge(train_base, train_label, on='user',how='left')
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x='age', hue='label', data=df)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('年龄', fontsize=12, fontproperties=font)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='provider', hue='label', data=df)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('运营商类型', fontsize=12, fontproperties=font)
plt.show()
```
